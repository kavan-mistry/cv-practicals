```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
```

```python
url = ["https://th.bing.com/th/id/OIP.kuLrtWwXW5nqy7UaVB2N8wHaHa?pid=ImgDet&rs=1"]

image1 = io.imread(url[0])
image1 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
cv2_imshow(image1)
```

```python
hsv_nemo = cv.cvtColor(image1, cv.COLOR_RGB2HSV)
light_white = (0,0,200)
dark_white = (145,60,255)
mask = cv.inRange(hsv_nemo, light_white, dark_white)
result = cv.bitwise_and(image1, image1, mask=mask)

cv2_imshow(result)
```
